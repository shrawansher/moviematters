### Load IMDB Data

```python
imdb = pd.read_csv('data/imdb_raw.csv')
imdb.head()
```

### Rename IMDB Columns

```python
def rename_imdb_cols(imdb):
    old_cols = imdb.columns
    print('\nCurrent Cols\n', old_cols)
    
    new_cols = ['imdb_index', 'imdb_id', 'imdb_primaryTitle', 'imdb_originalTitle', 'imdb_isAdult',
       'imdb_startYear', 'imdb_runtimeMinutes', 'imdb_genres', 'imdb_averageRating', 'imdb_numVotes',
       'imdb_directors']
    
    d= {}
    for i in range(len(old_cols)):
        d[old_cols[i]] = new_cols[i]
        
    print('\nCreated Dictionary & Renamed Columns\n', d)
    
    return imdb.rename(index=str, columns=d)    

```

```python
imdb = rename_imdb_cols(imdb)
```

```python
imdb.head()
```

### Merge TMDB and IMDB

```python
result = pd.merge(tmdb, imdb, on='imdb_id')
result = result.drop(['imdb_index'], axis = 1)
result.head()
```
